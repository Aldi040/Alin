#MATRIKS TRANSFORMASI
Transformasi geometri merupakan perubahan suatu bidang geometri yang meliputi posisi, besar dan bentuknya sendiri


##1. Refleksi(Reflection)
dalam geometri adalah transformasi di mana setiap titik pada bidang digeser atau dipindahkan menuju garis atau cermin dengan jarak yang sama, yaitu dua kali jarak titik tersebut ke cermin.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan array a (vektor yang akan direfleksi)
a = np.array([[2, 4]])

# Definisikan matriks refleksi
R = np.array([[-1, 0],
              [0, 1]])

# Hitung hasil refleksi
c = a @ R
print("Hasil refleksi matriks a pada sumbu y:", c)

# Visualisasi transformasi vektor sebelum dan sesudah refleksi
plt.figure(figsize=(20, 20))

# Vektor asli
plt.quiver(0, 0, a[0, 0], a[0, 1], angles='xy', scale_units='xy', scale=1, color='red', label='Vektor asli (a)')

# Vektor setelah refleksi
plt.quiver(0, 0, c[0, 0], c[0, 1], angles='xy', scale_units='xy', scale=1, color='purple', label='Vektor setelah refleksi (c)')

# Pengaturan plot
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('label x')
plt.ylabel('label y')
plt.title('Visualisasi Refleksi Vektor oleh Matriks')

# Refleksi visualisasi:
# Vektor asli (a) = [2, 4] ditampilkan dalam warna merah.
# Vektor setelah refleksi (c) = hasil refleksi a terhadap sumbu y ditampilkan dalam warna ungu.
# Refleksi ini mengubah arah vektor a terhadap sumbu y.

# Menampilkan grafik
plt.show()
```

### 2. Rotasi (Rotation)

- Rotasi adalah transformasi yang memutar vektor atau titik di sekitar asal koordinat (0,0) dalam ruang dua dimensi (2D) atau di sekitar sumbu tertentu dalam ruang tiga dimensi (3D).
- Untuk rotasi di ruang 2D dengan sudut \(theta\), matriks rotasinya adalah:
  \[
  R(theta) = \begin{bmatrix}
  \cos \theta & -\sin \theta \\
  \sin \theta & \cos \theta
  \end{bmatrix}
  \]
- Di ruang 3D, rotasi dapat dilakukan di sekitar sumbu x, y, atau z menggunakan matriks rotasi yang sesuai.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan array a (vektor yang akan dirotasi)
a = np.array([[2, 4]])

# Definisikan sudut rotasi (dalam derajat) dan konversi ke radian
theta = 90  # Sudut rotasi dalam derajat
theta_rad = np.radians(theta)  # Konversi ke radian

# Definisikan matriks rotasi
R = np.array([[np.cos(theta_rad), -np.sin(theta_rad)],
              [np.sin(theta_rad), np.cos(theta_rad)]])

# Hitung hasil rotasi
c = a @ R
print("Hasil rotasi matriks a dengan sudut", theta, "derajat adalah:", c)

# Visualisasi transformasi vektor sebelum dan sesudah rotasi
plt.figure(figsize=(20, 20))

# Vektor asli
plt.quiver(0, 0, a[0, 0], a[0, 1], angles='xy', scale_units='xy', scale=1, color='red', label='Vektor asli (a)')

# Vektor setelah rotasi
plt.quiver(0, 0, c[0, 0], c[0, 1], angles='xy', scale_units='xy', scale=1, color='purple', label='Vektor setelah rotasi (c)')

# Pengaturan plot
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('label x')
plt.ylabel('label y')
plt.title('Visualisasi Rotasi Vektor oleh Matriks')

# Refleksi visualisasi:
# Vektor asli (a) = [2, 4] ditampilkan dalam warna merah.
# Vektor setelah rotasi (c) = hasil rotasi a sebesar 90 derajat ditampilkan dalam warna ungu.
# Rotasi ini mengubah arah vektor a sesuai dengan sudut yang diberikan.

# Menampilkan grafik
plt.show()
```

### 3. Shear (Penggeseran)

- Shear adalah transformasi yang menggeser titik-titik dalam satu arah (x atau y) tergantung pada posisi mereka dalam arah yang lain.
- Contoh matriks shear dalam ruang 2D adalah:
   \begin{bmatrix}
  1 & k \\
  0 & 1
  \end{bmatrix}
  \
  di mana \( k \) adalah faktor shear.



```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan array a (vektor yang akan shear)
a = np.array([[2, 4]])

# Definisikan faktor shear untuk melakukan pergeseran matrik
k = 1.5  # faktor shear baru

# Definisikan matriks shear horizontal pada sumbu y
R = np.array([[1, k],
              [0, 1]])

# Hitung hasil pergeseran
c = a @ R
print("Hasil shear matriks a dengan faktor shear", k, "horizontal pada sumbu y:", c)

# Visualisasi transformasi vektor sebelum dan sesudah pergeseran
plt.figure(figsize=(20, 20))

# Vektor asli
plt.quiver(0, 0, a[0, 0], a[0, 1], angles='xy', scale_units='xy', scale=1, color='red', label='Vektor asli (a)')

# Vektor setelah pergeseran
plt.quiver(0, 0, c[0, 0], c[0, 1], angles='xy', scale_units='xy', scale=1, color='purple', label='Vektor setelah shear (c)')

# Pengaturan plot
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('label x')
plt.ylabel('label y')
plt.title('Visualisasi Shear Vektor oleh Matriks')

# Refleksi visualisasi:
# Vektor asli (a) = [2, 4] ditampilkan dalam warna merah.
# Vektor setelah pergeseran (c) = hasil shear a dalam warna ungu.
# Pergeseran ini mengubah arah vektor a sesuai dengan faktor shear 1.5.

# Menampilkan grafik
plt.show()
```

## 2. Skalasi (Scaling):
   - Skalasi mengubah ukuran objek dengan memperbesar atau memperkecilnya.
   - Matriks skalasi untuk ruang 2D dengan faktor skalasi \( s_x \) dan \( s_y \) adalah:
     \[
     S = \begin{bmatrix}
     s_x & 0 \\
     0 & s_y
     \end{bmatrix}
     \]
   - Untuk ruang 3D, faktor skalasi juga dapat diterapkan pada sumbu z.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan array a (vektor yang akan diskalasi)
a = np.array([[4, 2]])

# Definisikan faktor skalasi x dan faktor skalasi y untuk melakukan scaling matrik
fx = 1.5  # faktor skalasi x
fy = 2    # faktor skalasi y

# Definisikan matriks skalasi
R = np.array([[fx, 0],
              [0, fy]])

# Hitung hasil skalasi
c = a @ R
print("Hasil scaling matriks a dengan faktor x", fx, "dan faktor y", fy, "adalah", c)

# Visualisasi transformasi vektor sebelum dan sesudah skalasi
plt.figure(figsize=(10, 10))

# Vektor asli
plt.quiver(0, 0, a[0, 0], a[0, 1], angles='xy', scale_units='xy', scale=1, color='red', label='Vektor asli (a)', width=0.005)

# Vektor setelah scaling
plt.quiver(0, 0, c[0, 0], c[0, 1], angles='xy', scale_units='xy', scale=1, color='purple', label='Vektor setelah scaling (c)', width=0.005)

# Pengaturan plot
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('label x')
plt.ylabel('label y')
plt.title('Visualisasi Skalasi Vektor oleh Matriks')

# Skalasi visualisasi:
# Vektor asli (a) = [4, 2] ditampilkan dalam warna merah.
# Vektor setelah skalasi (c) = hasil skalasi a dengan faktor x 1.5 dan faktor y 2 ditampilkan dalam warna ungu.
# Skalasi ini mengubah ukuran vektor a sesuai dengan faktor yang diberikan.

# Menampilkan grafik
plt.show()
```
