##C. Sifat-Sifat Determinan

### 1. Determinan Matriks Identitas
Matriks identitas adalah matriks persegi dengan elemen 1 pada diagonal utama dan 0 pada posisi lainnya. Determinan dari matriks identitas adalah 1.

Misalkan $A$ dan $B$ adalah matriks identitas berukuran $n \times n$, maka

$$\text{Determinan } A = \det(A) = 1$$
$$\text{Determinan } B = \det(B) = 1$$


```python
import numpy as np

A = np.array([[1, 0], [0, 1]])  # matriks identitas
B = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])  # matriks identitas

# Determinan matriks A dan B
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)

print("Determinan A:", det_A)
print("Determinan B:", det_B)
```

### 2. Determinan Matriks Setelah Pertukaran Baris
Jika dua baris dari sebuah matriks ditukar, maka determinannya akan berubah tanda.

Misalkan $C$ adalah matriks awal dan $D$ adalah matriks yang diperoleh setelah menukar dua baris pada matriks $C$, maka

$$\text{Determinan } C = \det(C) = c$$
$$\text{Determinan } D = \det(D) = -c$$


```python
import numpy as np

C = np.array([[1, 2, 1], [3, 1, 2], [2, 1, 2]])  # matriks awal
D = np.array([[1, 2, 1], [2, 1, 2], [3, 1, 2]])  # merubah baris 2 dengan 3

# Determinan matriks C dan D
det_C = np.linalg.det(C)
det_D = np.linalg.det(D)

print("Determinan C:", det_C)
print("Determinan D:", det_D)
```

### 3. Baris Matriks Adalah Kelipatan dari Baris Lainnya
Jika suatu baris matriks adalah kelipatan dari baris lainnya, maka determinan dari matriks tersebut adalah nol.

Misalkan $E$ adalah matriks di mana baris ke-2 merupakan kelipatan dari baris ke-3, maka

$$\text{Determinan } E = \det(E) = 0$$


```python
import numpy as np

E = np.array([[1, 2, 1], [2, 4, 2], [2, 1, 1]])  # baris 2 merupakan kelipatan dari baris 3

# Determinan matriks E
det_E = np.linalg.det(E)

print("Determinan E:", det_E)
```

### 4. Kolom Matriks Adalah Kelipatan dari Kolom Lainnya
Jika suatu kolom matriks adalah kelipatan dari kolom lainnya, maka determinan dari matriks tersebut adalah nol.

Misalkan $F$ adalah matriks di mana kolom ke-3 merupakan kelipatan dari kolom ke-1, maka

$$\text{Determinan } F = \det(F) = 0$$


```python
import numpy as np

F = np.array([[1, 2, 1], [2, 4, 2], [2, 1, 2]])  # kolom 3 merupakan kelipatan dari kolom 1

# Determinan matriks F
det_F = np.linalg.det(F)

print("Determinan F:", det_F)
```

### 5. Matriks Adalah Kelipatan dari Matriks Lainnya
Jika suatu matriks merupakan kelipatan dari matriks lainnya, maka determinannya adalah kelipatan pangkat ordo dari matriks tersebut, dikalikan dengan determinan matriks kelipatannya.

Misalkan $H$ adalah matriks yang merupakan kelipatan dari matriks $C$, maka

$$\text{Determinan } H = \det(H) = k \cdot \det(C)$$


```python
import numpy as np

C = np.array([[1, 2, 1], [3, 1, 2], [2, 1, 2]])  # matriks C
H = np.array([[6, 2, 4], [2, 4, 2], [4, 2, 4]])  # matriks H merupakan kelipatan dari matriks C

# Determinan matriks C dan H
det_C = np.linalg.det(C)
det_H = np.linalg.det(H)

print("Determinan C:", det_C)
print("Determinan H:", det_H)
```

### 6. Matriks Adalah Hasil Penjumlahan dari Dua Matriks
Jika suatu matriks adalah hasil penjumlahan dari dua matriks, maka determinannya adalah hasil penjumlahan determinan matriks-matriks tersebut.

Misalkan $M$ adalah hasil penjumlahan dari matriks $K$ dan $L$, maka

$$\text{Determinan } M = \det(M) = \det(K) + \det(L)$$


```python
import numpy as np

K = np.array([[6, 4], [4, 1]])  # matriks K
L = np.array([[6, 8], [2, 1]])  # matriks L
M = K + L  # hasil penjumlahan matriks K dan L

# Determinan matriks K, L, dan M
det_K = np.linalg.det(K)
det_L = np.linalg.det(L)
det_M = np.linalg.det(M)

print("Determinan K:", det_K)
print("Determinan L:", det_L)
print("Determinan M:", det_M)
```

### 7. Matriks Adalah Segitiga Atas
Jika suatu matriks membentuk segitiga atas dengan nol di atas atau di bawah diagonal, maka determinannya adalah hasil perkalian diagonalnya.

Misalkan $N$ adalah matriks yang membentuk segitiga atas, maka

$$\text{Determinan } N = \det(N) = \text{hasil perkalian elemen diagonal}$$


```python
import numpy as np

N = np.array([[7, 5], [0, 1]])  # matriks N membentuk segitiga atas

# Determinan matriks N
det_N = np.linalg.det(N)

print("Determinan N:", det_N)
```

### 8. Hasil Perkalian Dua Matriks
Jika dua matriks dikalikan, determinan dari hasil matriks tersebut adalah hasil perkalian determinan matriks tersebut.

Misalkan $P$ adalah hasil perkalian dari matriks $O$ dan $Q$, maka

$$\text{Determinan } P = \det(P) = \det(O) \cdot \det(Q)$$


```python
import numpy as np

O = np.array([[1, 3], [0, 2]])  # matriks O (2x2)
Q = np.array([[2, 1], [1, 1]])  # matriks Q (2x2)

# Hasil perkalian matriks O dan Q
P = np.dot(O, Q)

# Determinan matriks O, Q, dan P
det_O = np.linalg.det(O)
det_Q = np.linalg.det(Q)
det_P = np.linalg.det(P)

print("Determinan O:", det_O)
print("Determinan Q:", det_Q)
print("Determinan P:", det_P)
```

### 15. Determinan Matriks Transpose
Determinan dari matriks transpose adalah sama dengan determinan dari matriks aslinya.

Misalkan $A$ adalah suatu matriks, maka

$$\det(A^T) = \det(A)$$


```python
import numpy as np

A = np.array([[1, 2], [3, 4]])  # matriks A

# Determinan matriks A
det_A = np.linalg.det(A)

# Determinan matriks transpose dari A
det_A_transpose = np.linalg.det(A.T)

print("Determinan A:", det_A)
print("Determinan transpose dari A:", det_A_transpose)
```
