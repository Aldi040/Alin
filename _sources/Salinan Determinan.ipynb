#Sifat Determinan Matriks

##A. macam- macam Contoh matriks

```python
import numpy as np

A = np.array([[1, 0], [0, 1]])  # matriks identitas
B = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])  # matriks identitas
C = np.array([[1, 2, 1], [3, 1, 2], [2, 1, 2]])  # matriks identitas
D = np.array([[1, 2, 1], [2, 1, 2], [3, 1, 2]])  # merubah baris 2 dengan 3
E = np.array([[1, 2, 1], [2, 4, 2], [2, 1, 1]])  # matriks dengan mengubah baris 2 dengan 3
F = np.array([[1, 2, 1], [2, 4, 2], [2, 1, 2]])  # matriks dengan kolom yang merupakan kelipatan dari baris lainnya
G = np.array([[1, 2, 1], [2, 4, 2], [2, 4, 3]])  # matriks dengan kolom yang merupakan kelipatan dari kolom lainnya
H = np.array([[6, 2, 4], [2, 4, 2], [4, 2, 4]])  # merubah baris 2 dengan 3
I = np.array([[3, 4], [2, 1]])  # matriks identitas
J = np.array([[6, 8], [4, 2]])  # matriks identitas
K = np.array([[6, 4], [4, 1]])  # kolom sama
L = np.array([[6, 8], [2, 1]])  # baris sama
M = np.array([[4, 1], [2, 1]])  # baris sama
N = np.array([[7, 5], [2, 1]])  # baris sama
O = np.array([[1, 3, 2], [0, 2, 2], [0, 0, 1]])  # merubah baris 2 dengan 3
P = np.array([[1, 0, 0], [1, 2, 0], [1, 2, 1]])  # merubah baris 2 dengan 3
Q = np.array([[2, 1], [1, 1]])  # baris sama
R = np.array([[1, 1], [1, 1]])  # baris sama
S = np.array([[3, 2], [3, 2]])  # baris sama
T = np.array([[3, 1, 2],[2, 1, 2], [4, 2, 1]])
U = np.array([[]])
#Cetak matrks
print("Cetak Matriks:")
print("Matriks A:")
print(A)
print()

print("Matriks B:")
print(B)
print()

print("Matriks C:")
print(C)
print()

print("Matriks D:")
print(D)
print()

print("Matriks E:")
print(E)
print()

print("Matriks F:")
print(F)
print()

print("Matriks G:")
print(G)
print()

print("Matriks H:")
print(H)
print()

print("Matriks I:")
print(I)
print()

print("Matriks J:")
print(J)
print()

print("Matriks K:")
print(K)
print()

print("Matriks L:")
print(L)
print()

print("Matriks M:")
print(M)
print()

print("Matriks N:")
print(N)
print()

print("Matriks O:")
print(O)
print()

print("Matriks P:")
print(P)
print()

print("Matriks Q:")
print(Q)
print()

print("Matriks R:")
print(R)
print()

print("Matriks S:")
print(S)

print("matriks T:")
print(T)
```

##B. menghitung determinan

```python
# Hitung determinan masing-masing matriks
detA = np.linalg.det(A)
detB = np.linalg.det(B)
detC = np.linalg.det(C)
detD = np.linalg.det(D)
detE = np.linalg.det(E)
detF = np.linalg.det(F)
detG = np.linalg.det(G)
detH = np.linalg.det(H)
detI = np.linalg.det(I)
detJ = np.linalg.det(J)
detK = np.linalg.det(K)
detL = np.linalg.det(L)
detM = np.linalg.det(M)
detN = np.linalg.det(N)
detO = np.linalg.det(O)
detP = np.linalg.det(P)
detQ = np.linalg.det(Q)
detR = np.linalg.det(R)
detS = np.linalg.det(S)
detT = np.linalg.det(T)
```

##C. Sifat-Sifat Determinan

```python
#Sifat-Sifat determinan
print("Sifat-Sifat determinan")
print("determinan Matriks")
# Matriks identitas memiliki determinan 1.
print("Determinan A (matriks identitas):", int(detA))
print("Determinan B (matriks identitas):", int(detB))
# matriks C
print("Determinan C matriks C: ",int(detC))
# matriks c dengan merubah baris akan mengubah tanda dari determinan
print("Determinan D (matriks c dengan merubah baris akan mengubah tanda dari determinan):",int(detD))
# Jika suatu baris matriks adalah kelipatan dari baris lainnya, determinannya akan menjadi nol.
print("Determinan E (baris 2 merupakan kelipatan dari baris 3):", int(detE))

# Jika suatu kolom matriks adalah kelipatan dari kolom lainnya, determinannya juga akan menjadi nol.
print("Determinan F (kolom 3 merupakan kelipatan dari kolom 1):", int(detF))
print("Determinan G (kolom 3 merupakan kelipatan dari kolom 2):", int(detG))

# Untuk matriks yang merupakan kelipatan dari matriks lainnya, determinannya adalah kelipatan pangkat ordo dari matriks tersebut, dikalikan dengan determinan matriks kelipatannya.
print("Determinan H (matriks H merupakan kelipatan dari matriks C):", int(detH))
# Jika suatu matriks adalah hasil penjumlahan dari dua matriks, determinannya adalah hasil penjumlahan determinan matriks-matriks tersebut.
print("Determinan M (hasil penjumlahan matriks K dan L):", detM)

# Jika suatu matriks membentuk segitiga atas dengan nol di atas atau di bawah diagonal, determinannya adalah hasil perkalian diagonalnya.
print("Determinan N (matriks N membentuk segitiga atas):", detN)

# Jika dua matriks dikalikan, determinan dari hasil matriks adalah hasil perkalian determinan matriks tersebut.
print("Determinan P (hasil perkalian matriks O dan Q):", detP)
print("Determinan Q (matriks Q):", detQ)
print("determinan T:",detT)
```
